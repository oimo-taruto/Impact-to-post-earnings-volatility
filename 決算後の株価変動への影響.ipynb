```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display,Markdown

def get_earnings_impact(ticker_symbol):
  try:
    display(Markdown(f"---"))
    display(Markdown(f"## 📈分析レポート：{ticker_symbol}"))

    ticker_obj = yf.Ticker(ticker_symbol)

    earnings_dates = ticker_obj.get_earnings_dates()
    earnings_dates_list = [date.date() for date in earnings_dates.index]

    all_data = yf.download(tickers=ticker_symbol, period='max',auto_adjust=False,progress=False)

    if isinstance(all_data.columns, pd.MultiIndex):
      all_data.columns = all_data.columns.get_level_values(0)

    df=all_data[['Open','High','Low','Close','Adj Close','Volume']].copy()
    df = df.reset_index()
    df['Nextday_Open'] = df['Open'].shift(-1)
    df['Nextday_Close'] = df['Close'].shift(-1)
    df['Volatility'] = (df['Nextday_Close'] - df['Close']) * 100 / df['Close']
    df['Volatility_open'] = (df['Nextday_Open'] - df['Close']) * 100 / df['Close']
    df['Volatility_before_10d']=(df['Close']-df['Close'].shift(-10))*100/df['Close'].shift(-10)
    df['Volatility_after_10d']=(df['Close'].shift(10)-df['Close'])*100/df['Close']

    earnings_dates_dt = pd.to_datetime(earnings_dates_list)
    earnings_data = df[df['Date'].isin(earnings_dates_dt)]

    earnings_volatility = earnings_data['Volatility']

    df_ticker_symbol = pd.DataFrame(earnings_dates)

    earnings_data = earnings_data.reset_index()
    df_ticker_symbol = df_ticker_symbol.reset_index()

    earnings_data['Date'] = pd.to_datetime(earnings_data['Date']).dt.normalize().dt.tz_localize(None)
    df_ticker_symbol['Earnings Date'] = pd.to_datetime(df_ticker_symbol['Earnings Date']).dt.normalize().dt.tz_localize(None)

    merged_df = pd.merge(earnings_data, df_ticker_symbol, left_on='Date', right_on='Earnings Date')
    merged_df = merged_df.drop(columns=['Earnings Date'])
    merged_df.replace([np.inf,-np.inf],np.nan,inplace=True)
    merged_df.dropna(inplace=True)

    merged_df['ticker_symbol'] = ticker_symbol

    target_cols=['Surprise(%)','EPS Estimate','Volatility','Volatility_open','Volatility_after_10d','Volatility_before_10d']
    df_corr=merged_df[target_cols].corr()
    display(df_corr.style.background_gradient(cmap='coolwarm',vmin=-1,vmax=1).format("{:.3f}"))

    import statsmodels.api as sm

    x=merged_df['Surprise(%)']
    y=merged_df['Volatility_open']
    y2=merged_df['Volatility_after_10d']
    z=merged_df['Volatility_before_10d']
    x_s=(x-x.mean())/x.std()
    y_s=(y-y.mean())/y.std()
    y2_s=(y2-y2.mean())/y2.std()
    z_s=(z-z.mean())/z.std()

    X=merged_df[['Surprise(%)','Volatility_before_10d']]
    X_s=(X-X.mean())/X.std()
    X_s=sm.add_constant(X_s)
    model=sm.OLS(y_s,X_s).fit(cov_type='HC3')
    model2=sm.OLS(y2_s,X_s).fit(cov_type='HC3')

    from statsmodels.iolib.summary2 import summary_col
    import io

    res=summary_col([model,model2],stars=True,float_format='%0.3f',model_names=['Open Volatility','After 10days Volatility'])
    res_df=pd.read_html(io.StringIO(res.as_html()),header=0,index_col=0)[0]
    res_df.index=res_df.index.fillna('')
    res_df=res_df.fillna('')
    display(res_df)

    display(model.summary().tables[0])
    display(model.summary().tables[1])
    display(model2.summary().tables[0])
    display(model2.summary().tables[1])

    return merged_df

  except Exception as e:
    print(f"An error occured at {ticker_symbol}:{e}")
    return e

target_tickers=['AAPL','MSFT','GOOGL','AMZN','META','MCD']
results=[]
for s in target_tickers:
  results.append(get_earnings_impact(s))

final_all=pd.concat(results)

import seaborn as sns

sns.set_theme(style="whitegrid")
sns.set_context("paper",font_scale=1.2)
g1=sns.FacetGrid(final_all,col="ticker_symbol",col_wrap=3,height=4,sharex=False,sharey=False)
g1.map(sns.regplot,"Volatility_before_10d","Volatility_open",scatter_kws={'alpha':0.5},line_kws={'color':'blue'})
g1.set_axis_labels("Pre-earnings Volatility(10days)","Post-eranings Volatility(open)")
g1.fig.suptitle("Anomary Verification: Pre-earnings Volatility VS Post-earnings Volatility",y=1.03,fontsize=16,fontweight='bold')
plt.show()

g2=sns.FacetGrid(final_all,col="ticker_symbol",col_wrap=3,height=4,sharex=False,sharey=False)
g2.map(sns.regplot,"Surprise(%)","Volatility_open",scatter_kws={'alpha':0.5},line_kws={'color':'green'})
g2.set_axis_labels("EPS Surprise(%)","Post-eranings Volatility(open)")
g2.fig.suptitle("Influence Verification: EPS Surprise VS Post-earnings Volatility",y=1.03,fontsize=16,fontweight='bold')
plt.show()
```
